Importy i wstępna konfiguracja

```python
import matplotlib.pyplot as plt
import csv

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100
```

Deklaracja słowników kategorii

```python
categories = {
    'event': {},
    'event_hyponym': {},
    'entity': {},
    'entity_hyponym': {},
    'out_in': {},
    'location': {},
}
dictionary = {}
```

Odczyt kategorii obrazów

```python
classification_filename = '../AIDe_ANNOTATED_DESCRIPTIONS/classification.csv'

keys = list(categories.keys())

with open(classification_filename, newline='', encoding='utf-8') as f:
    reader = csv.reader(f, delimiter='\t')
    img_names = []
    next(reader)
    for row in reader:
        for i in range(2,8):
            words=row[i].split(',')
            for word in words:
                if word in categories[keys[i-2]]:
                    categories[keys[i-2]][word] = categories[keys[i-2]][word] + 1
                else:
                    categories[keys[i-2]][word] = 1
```

```python
captions_filename = '../AIDe_ANNOTATED_DESCRIPTIONS/captions.csv'

with open(captions_filename, newline='', encoding='utf-8') as f:
    reader = csv.reader(f, delimiter='\t')
    img_names = []
    next(reader)
    for row in reader:
        for i in range(2,4):
            words=row[i].replace(',', '').split(' ')
            for word in words:
                if len(word) >= 2:
                    if word in dictionary:
                        dictionary[word] = dictionary[word] + 1
                    else:
                        dictionary[word] = 1
```

Wykresy słupkowe przedstawicieli danych kategorii

```python
keys = list(categories.keys())

for key in keys:
    sorted_dict = dict(sorted(categories[key].items(),key=lambda item: item[1],reverse=True))

    names = list(sorted_dict.keys())
    values = list(sorted_dict.values())
    
    plt.bar(names, values)
    plt.title(key)
    plt.xticks(rotation=90)
    plt.show()
```
